```python
import pandas as pd 
import numpy as np
#https://www.ncei.noaa.gov/pub/data/cirs/climdiv/climdiv-tminst-v1.0.0-20230206
#https://www.ncei.noaa.gov/pub/data/cirs/climdiv/climdiv-tmaxst-v1.0.0-20230206

#urls = ["https://www.ncei.noaa.gov/pub/data/cirs/climdiv/climdiv-tmaxst-v1.0.0-20230206", "", "", ]

years = [str(x) for x in range(2000,2023)]

states = [    'ALABAMA',    'ALASKA',    'ARIZONA',    'ARKANSAS',    'CALIFORNIA',    'COLORADO',    'CONNECTICUT',    'DELAWARE',    'FLORIDA',    'GEORGIA',    'HAWAII',    'IDAHO',    'ILLINOIS',    'INDIANA',    'IOWA',    'KANSAS',    'KENTUCKY',    'LOUISIANA',    'MAINE',    'MARYLAND',    'MASSACHUSETTS',    'MICHIGAN',    'MINNESOTA',    'MISSISSIPPI',    'MISSOURI',    'MONTANA',    'NEBRASKA',    'NEVADA',    'NEW HAMPSHIRE',    'NEW JERSEY',    'NEW MEXICO',    'NEW YORK',    'NORTH CAROLINA',    'NORTH DAKOTA',    'OHIO',    'OKLAHOMA',    'OREGON',    'PENNSYLVANIA',    'RHODE ISLAND',    'SOUTH CAROLINA',    'SOUTH DAKOTA',    'TENNESSEE',    'TEXAS',    'UTAH',    'VERMONT',    'VIRGINIA',    'WASHINGTON',    'WEST VIRGINIA',    'WISCONSIN',    'WYOMING']

# Create a list of all years
years = range(2000, 2023)

# Create a list of all months
months = range(1, 13)

# Create an empty list to store the data
keys = []

# Loop over all states, years, and months to create the data list
for state in states:
    for year in years:
        for month in months:
            keys.append((state, year, month))

# Create a pandas dataframe from the data list
df = pd.DataFrame(keys, columns=['state', 'year', 'month'])

df.head()
```
